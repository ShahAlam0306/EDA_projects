```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#rename for clarity and better understanding for myself
df.columns = df.columns.str.lower()
df = df.rename(columns = {'customer id': 'cust_id',
                         'item purchased':'i_purchased',
                         'discount applied':'d_applied',
                          'subscription status':'subs_status',
                          'purchase amount (usd)':'purchased_amt',
                          'promo code used':'promo_code_used',
                          'previous purchases':'prev_purchase',
                          'payment method' : 'pay_method',
                          'frequency of purchases':'freq_of_puchases',
                         'review rating' : 'review_rating',
                          'shipping type' : 'shipping_type'
                         
                         })
df
```

```python
df.columns
```

```python
#creating a new age_group columns
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q = 4,labels = labels)
```

```python
df[['age','age_group']].head()
```

```python
#create columns purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly' :7,
    'Monthly': 30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months': 90   
}
df['purchase_freq_days'] = df['freq_of_puchases'].map(frequency_mapping)
```

```python
print("\nFirst 10 rows of the data and last 10 rows of data: ")
print(df.head(10))
print(df.tail(10))
```

```python
print('\nMissing values in each column: ')
print(df.isnull().sum(),'\n')
```

```python
df['Total Recovered'] = df['Total Recovered'].fillna(0)
df['Total Deaths'] = df['Total Deaths'].fillna(0)
df['Total Cases'] = df['Total Cases'].fillna(0)
df['Total Test'] = df['Total Test'].fillna(0)
df['Active Cases'] = df['Active Cases'].fillna(0)
df.dropna()
print('DataFrame after handling missing values: ')
print(df,'\n')
```

```python
print('Number of duplicate rows and columns: ',df.duplicated().sum())
df_no_duplicates = df.drop_duplicates()
```

```python
print('Since no duplicates rows are avalaible so our data is like this:-> ')
print(df_no_duplicates)
```

```python
print("Final DataFrame info using pandas: ")
print(df_no_duplicates.info())
```

```python
print("Now we want to find maximum deaths across countries: ")
df['Total Percent'] = df['Total Deaths'] * 100.00 / df['Total Cases']
print(df.head(14))
print(df.tail(14),'\n')
```

```python
df.info()
```

```python
# Renaming columns for clarity 
df = df.rename(columns = {
    'Serial Number' : "No.",
    'Total Cases' : 'T_cases',
    'Total Deaths' : 'T_deaths',
    'Total Recovered' : 'T_recovered',
    'Total Test' : 'T_test',
    'Total Percent' : 'T_percent',
    'Population' : 'Pop'
})
```

```python
print("DataFrame Info(data types and non-null: ")
print(df.info())
print('\n')
```

```python
#Show high percent of maximum deaths across countries
print(df[df['T_percent'] == df['T_percent'].max()])


```
